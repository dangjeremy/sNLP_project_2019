# NLP Assignment 3 - check BERT hyperparameters

This notebook was compiled so that we could identify the best sets of hyperparamters for each training set size. 

## Imports

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```
from google.colab import drive
drive.mount('/content/gdrive')
```

## Hyperparameters and path

```
learning_rates = [5e-5, 3e-5, 2e-5]
batch_sizes = [16, 32]
nrepeats = 5
```

```
#Select the path contained the datasets
CVpath = '/content/gdrive/My Drive/SARC pol/CVruns/'
```

## 100% project training set

```
CV_100_means = pd.DataFrame()
CV_100_std = pd.DataFrame()
```

```
for b in batch_sizes:
  for lr in learning_rates:
    comb_array = np.zeros((5,5))
    for n in range(0, nrepeats):
        with open(CVpath+f"results_train100_{lr}_{b}_{n}.pickle", 'rb') as handle:
          temp = pickle.load(handle)
          
          #Get the validation accuracy column
          v = temp["validation accuracy"]
          
          #Store it in an array
          comb_array[:,n] = v
    
    #Get the mean and std for each epoch over the 5 repeats
    comb_mean = np.mean(comb_array, axis=1)
    comb_std = np.std(comb_array, axis=1)
    
    #Store them in dataframes
    CV_100_means[f"(bs:{b}, lr:{lr})"] = comb_mean
    CV_100_std[f"(bs:{b}, lr:{lr})"] = comb_std 
```
